Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans-- A scatter plot is a type of chart used to display the relationship or correlation between two numerical variables. It is a graphical representation of the correlation between two variables, where each point on the plot represents a single observation in the dataset.

In a scatter plot, the values of the first variable are plotted on the horizontal axis, while the values of the second variable are plotted on the vertical axis. Each data point is represented by a dot on the plot, with the position of the dot indicating the value of both variables for that observation.

Scatter plots are useful for identifying patterns and relationships between variables. For example, if the dots on the plot form a straight line, it indicates a strong linear relationship between the two variables. On the other hand, if the dots are scattered all over the plot, it indicates a weak or no relationship between the two variables.

Scatter plots can also be used to identify outliers, which are observations that are significantly different from the other observations in the dataset. Outliers may indicate errors in data collection or data processing, or they may represent important data points that require further investigation.

```python
import matplotlib.pyplot as mp
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
mp.title("Graph")
mp.xlabel("X-axis")
mp.ylabel("Y-axis")
mp.scatter(x,y)
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans--> The subplot() function in Matplotlib is used to create multiple plots within a single figure. It takes three arguments: the number of rows, the number of columns, and the index of the current plot. By specifying the index of the current plot, we can create multiple plots that share the same figure.

Here are four line plots using the subplot() function:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1 in the top-left subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2 in the top-right subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3 in the bottom-left subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4 in the bottom-right subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a title to the entire figure
fig.suptitle('Four Line Plots')

# Show the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

Ans--> A bar plot is a type of chart used to display and compare categorical data. It represents data with rectangular bars where the height or length of the bar corresponds to the value of the variable being displayed. Each bar represents a category or group, and the bars can be arranged vertically or horizontally depending on the type of data being displayed.

Bar plots are used to compare the values of different categories or groups in a dataset. They are particularly useful for displaying discrete data that is not continuous, such as counts or frequencies of categorical variables. Bar plots can also be used to show the relationship between two or more categorical variables.

For example, a bar plot could be used to display the sales figures for different products in a store, with each product represented by a separate bar. The height of each bar would correspond to the sales figure for that product, making it easy to compare the sales figures for different products.

Bar plots can also be used to show the distribution of categorical data, such as the number of people in different age groups or the frequency of different types of vehicles on the road. In this case, the bars would represent the different categories or groups, and the height or length of the bar would represent the frequency or count of each category or group.

Overall, bar plots are a simple yet powerful tool for displaying and comparing categorical data, making them a popular choice in data visualization.


plotting bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans--> A box plot, also known as a box-and-whisker plot, is a type of chart used to display the distribution of a dataset. It is commonly used to summarize and compare the distribution of a variable across different categories or groups.

A box plot displays a summary of the distribution of the data by dividing it into quartiles. The box represents the middle 50% of the data, with the median represented by a line within the box. The "whiskers" or lines extending from the box show the range of the data, excluding outliers, which are plotted as individual points beyond the whiskers.

Box plots are used to visualize the spread of a dataset and identify any outliers. They are particularly useful for comparing the distribution of a variable across different categories or groups. For example, a box plot could be used to compare the distribution of test scores for different schools, with each school represented by a separate box plot. The height or length of the box plot would represent the middle 50% of the test scores for each school, making it easy to compare the distribution of scores across different schools.

Box plots can also be used to identify skewness or outliers in a dataset. If a box plot is skewed, it may indicate that the data is not normally distributed, and further analysis may be necessary to understand the underlying pattern. Similarly, if there are outliers in a box plot, it may be important to investigate these data points further to determine if they represent valid data or errors in the data collection process.

Overall, box plots are a powerful tool for visualizing and summarizing the distribution of a dataset and comparing the distribution of a variable across different categories or groups.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis object
fig, ax = plt.subplots()

# Create a box plot for the two datasets
ax.boxplot([box1, box2])

# Set the title and axis labels
ax.set_title('Box plot of box1 and box2')
ax.set_xlabel('Box')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```
